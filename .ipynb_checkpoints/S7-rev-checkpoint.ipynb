## 1.  Line Plot

In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and aout three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti.

```python
haiti = df_can.loc['Haiti', years]
```

```python
haiti.plot()
```

```python
haiti.index = haiti.index.map(int)
```

```python
haiti.plot(kind='line')

plt.title('Imigration from Haiti')
plt.ylabel('Number of Imigrants')
plt.xlabel('Years')

plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

## 3 Histogram

What is the frequency distribution of the number (population) of new immigrants from the various countries to Canada in 2013?

```python
df_can['2013'].head()
```

```python
count, bin_edges = np.histogram(df_can['2013'])
```

```python
df_can['2013'].plot(kind='hist', figsize=(14, 9))

plt.title('Histogram of Imigrations from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Imigrants')

plt.show()
```

```python
df_can['2013'].plot(kind='hist', figsize=(14, 9), xticks=bin_edges)

plt.title('Histogram of Imigrations from 195 Countries in 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Imigrants')

plt.show()
```

```python
df_can.loc[['Denmark', 'Norway', 'Sweden'], years]
```

```python
df_t = df_can.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
```

```python
df_t.plot(kind='hist', figsize=(20, 10))

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Imigrants')

plt.show()
```

- increase the bin size to 15 by passing in bins parameter
- set transparency to 60% by passing in alpha paramemter
- label the x-axis by passing in x-label paramater
- change the colors of the plots by passing in color parameter

```python
count, bin_edges = np.histogram(df_t, 15)
```

```python
df_t.plot(kind = 'hist', 
          figsize = (20, 10),
          bins = 15,
          alpha = 0.6,
          xticks = bin_edges,
          color = ['coral', 'darkslateblue', 'mediumseagreen']
         )

plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Countries')
plt.xlabel('Number of Imigrants')

plt.show()
```

## 4 Bar Chart

```python
df_Iceland = df_can.loc['Iceland', years]
```

## 5 Pie Chart

```python
df_continents = df_can.groupby('Continent', axis=0).sum()
```

```python
df_continents.head()
```

```python
df_continents['Total'].plot(kind='pie',
                           figsize=(20, 10),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True
                           )

plt.title('Imigrationg to Canada by Continents (1980 - 2013)')
plt.axis('equal')

plt.show()
```

```python
colors_list = ['aqua', 'lawngreen', 'blueviolet', 'gold', 'pink', 'red']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['Total'].plot(kind='pie',
                           figsize=(20, 10),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            colors=colors_list,
                            labels=None,
                            pctdistance=1.12,
                            explode=explode_list
                           )

plt.title('Imigrationg to Canada by Continents (1980 - 2013)', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

```python
colors_list = ['aqua', 'yellowgreen', 'blueviolet', 'gold', 'pink', 'red']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1]

df_continents['2013'].plot(kind='pie',
                           figsize=(20, 10),
                            autopct='%1.1f%%',
                            startangle=90,
                            shadow=True,
                            colors=colors_list,
                            labels=None,
                            pctdistance=1.12,
                            explode=explode_list
                           )

plt.title('Imigrationg to Canada by Continents in 2013', y=1.12)
plt.axis('equal')

plt.legend(labels=df_continents.index, loc='upper left')

plt.show()
```

## 7 Sub Plot

```python
fig = plt.figure() # create a canvas

ax0 = fig.add_subplot(1, 2, 1) # create worksheet 1
ax1 = fig.add_subplot(1, 2, 2) # # create worksheet 1

# Create plot in worksheet 1
df_CI.plot(kind='box', 
           figsize=(20, 10),
           color='darkorange',
           vert=False,
           ax=ax0
          )

ax0.set_title('Box plot of China and India Imigrants between 1980 - 2013')
ax0.set_ylabel('Number of Imigrants')
ax0.set_xlabel('Countries')

# Create plot in worksheet 2
df_CI.plot(kind='line', 
           figsize=(20, 10),
           ax=ax1
          )

ax1.set_title('Box plot of China and India Imigrants between 1980 - 2013')
ax1.set_ylabel('Number of Imigrants')
ax1.set_xlabel('Years')

plt.show()
```
